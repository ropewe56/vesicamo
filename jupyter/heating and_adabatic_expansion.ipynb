# Some Thermodymanics

## First Law of Thermdynamics
$dU = \delta Q + \delta W$

$\delta Q$ - heat

$\delta W = - p dV$ - mechanical work

**Internal Energy**
$U = U(V,T)$

$dU = \left(\dfrac{\partial U}{\partial V}\right)_T + \left(\dfrac{\partial U}{\partial T}\right)_V$

**Constant volume heat capacity**
$\left(\dfrac{\partial U}{\partial T}\right)_V = C_V$

$\delta Q = C_v dT + p dV + \left(\dfrac{\partial U}{\partial V}\right)_T dV$

## Ideal Gas Law

$p V_M = R T$

$p = \dfrac{L}{V_M} \dfrac{R}{L} T = n k_B T$

$\left(\dfrac{\partial U}{\partial V}\right)_T = 0$ 

### Constant Pressure: 

$V = V(T)$

$dV = \left(\dfrac{\partial V}{\partial T}\right)_p dT$

$\left(\dfrac{\partial \delta Q}{\partial T}\right)_p$ = $C_p$ = $C_V + p \left(\dfrac{\partial V}{\partial T}\right)_p + \left(\dfrac{\partial U}{\partial V}\right)_T$ 

$C_p = C_V + p \left(\dfrac{\partial V}{\partial T}\right)_p = p \dfrac{R}{p} = R$

**Constant pressure heat capacity**

$C_p - C_V = R$

### Adiabatic Change

$C_v dT + p dV + \left(\dfrac{\partial U}{\partial V}\right)_T dV = 0$

$\left(\dfrac{dT}{dV}\right)_{ad} =$ $-\dfrac{p+\left(\dfrac{\partial U}{\partial V}\right)_T}{C_V}$

$T + p dV + \left(ft(\dfrac{dT}{dV}\right)_{ad} =$ $-\dfrac{C_p - C_v}{C_V}$ $\left(\dfrac{\partial T}{\partial V}\right)_p$

**Ideal Gas** 

$\left(\dfrac{\partial T}{\partial V}\right)_p$ = $\dfrac{p}{R} = \dfrac{T}{V}$

$\dfrac{C_p}{C_V} = \kappa$ - adiabatic coefficient

$\left(\dfrac{dT}{dV}\right)_{ad} =$ $ - \left(\kappa - 1\right) \dfrac{T}{V}$

$d \ln T =  - \left(\kappa - 1\right)  d \ln V$

$\dfrac{T_0}{T} =  \dfrac{V^{\left(\kappa - 1\right)}}{V_0^{\left(\kappa - 1\right)}}$

$T_0 V_0^{\left(\kappa - 1\right)} =  T V^{\left(\kappa - 1\right)}$

(1) $T V^{\left(\kappa - 1\right)} = const.$

(2) $p V^{\kappa} = const.$


### Self Compression

Two steps:
1. Heating of a contant volume
    * Intial state: $T_0, p_0, V_0$
    * Temperature increase: $\Delta T = C_V \delta Q$
    * End state: $T_1, p_1, V_0$
2. Expansion of this volume
    * Intial state: $T_1, p_1, V_0$
    * Expansion from $p_1$ to $p_0$
    * End state: $T_2, p_0, V_2$

1. Heating

(3) $p_1 = R \dfrac{T_1}{V_0}$

Temperature increase

(4) $T_1 = T_0 +  C_V \delta Q$

2. Aiabatic expansion

with (2) and (3)

(5) $R \dfrac{T_1}{V_0} V_0^{\kappa} = p_0 V_2^{\kappa}$

with (5) and (4)

(6) $\dfrac{R}{p_0} \dfrac{T_0 +  C_V \delta Q}{V_0} V_0^{\kappa} = V_2^{\kappa}$

with $R/(p_0 V_0) = 1/T_0$

(7) $\dfrac{T_0 +  C_V \delta Q}{T_0} V_0^{\kappa} = V_2^{\kappa}$

with (1) 

(8) $T_1 V_0^{\left(\kappa - 1\right)} = T_2 V_2^{\left(\kappa - 1\right)}$ 

(9) $V_2^{\kappa} = \left(\dfrac{T_0}{T_1}\right)^{\kappa/(\kappa - 1)} V_0^{\kappa}$ 

with (9) and (7): 

$\left(\dfrac{T_0}{T_2}\right)^{\kappa/(\kappa - 1)} = \dfrac{T_0 +  C_V \delta Q}{T_0}$ 

$\dfrac{T_2}{T_0} = \left(\dfrac{T_0 +  C_V \delta Q}{T_0}\right)^{(\kappa - 1)/\kappa}$

```python
# Two atomic gas
k = 5/2
# Initial temperature
T0 = 280.0
# Temperature increase due to heating 
dT = 1.0
# Temperature after expansion
T2 = ((T0 + dT)/T0)**((k-1)/k) * T0
# Relative temperature increase after expansion
print("(T_2 - T_0)/dT = %4.2f" % ((T2 - T0)/dT))
```
